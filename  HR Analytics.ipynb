###### Data Visualisation

###### 1. Correlation map for  all numeric values

```python
# Plot heatmap of `correlation` using the 'inferno' colormap
plt.figure(figsize=(16,15))
corr_map = df.select_dtypes(include=['number']).corr()
sns.heatmap(corr_map,xticklabels=corr_map.columns, yticklabels=corr_map.columns,cmap='inferno', annot=True);
plt.title('Correlation Map for Numeric Variables', fontsize=30)
plt.show()
```

###### 2.Gender

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Gender', hue='Attrition')
plt.title('Attrition by Gender')
plt.show()
```

###### 3.JobRole

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='JobRole', hue='Attrition',color='')
plt.title('Attrition by Job Role')
plt.xticks(rotation=90)
plt.show()
```

###### 4.Age

```python
df.Age.describe()
```

```python
print(df.Age.skew())
sns.histplot(x="Age", data=df, binwidth=5, kde=True)
```

###### 5.Relationship between Overtime and Age

```python
sns.boxplot(data=df,x='OverTime',y='Age')
plt.title('Relationship Between Overtime and Age')
plt.show()
     
```

###### 6.Education Level

```python
plt.figure(figsize=(22, 10))
sns.barplot(x='Education', y='Attrition', data=df,)
plt.title('Attrition by Education Level', fontsize=30)
plt.xlabel('Education Level', fontsize=20)
plt.ylabel('Attrition Rate', fontsize=20)
plt.show()
```

###### 7.Business Travel

```python
custom_palette = ['#EE4266', '#824D74', '#401F71'] 
plt.figure(figsize=(8 , 8))
business_travel_counts = df['BusinessTravel'].value_counts()


plt.pie(business_travel_counts, labels=business_travel_counts.index,
        autopct=lambda p: f'{p:.1f}% ({int(p * sum(business_travel_counts)/100)})',
        colors=custom_palette)  
plt.title('Distribution of Business Travel', fontsize=30)
plt.show()
```

###### 8.Total Working Years

```python
sns.histplot(data=df,x='TotalWorkingYears',bins=10,kde=True)
plt.title('Total Working Years')
plt.show()
```

###### 9.Department

```python
Department_counts = df['Department'].value_counts()
department_colors = ['#EE4266', '#824D74', '#401F71'] 
plt.figure(figsize=(8, 8))
plt.pie(Department_counts, labels=Department_counts.index, autopct=lambda p: f'{p:.1f}% ({int(p)})', colors=department_colors)
plt.title('Distribution of Department\n', fontsize=30)
plt.axis('equal')
plt.show()
```

###### 10.Marital Status

```python
sns.countplot(data=df, x="MaritalStatus")
plt.title('Marital Status')
plt.show()
```

###### 11.Education Field

```python
custom_palette = ['#EE4266', '#824D74', '#401F71'] 
plt.figure(figsize=(22, 10))
plt.subplot(1, 2, 2)
sns.countplot(y='JobRole', data=df, order=df['JobRole'].value_counts().index, palette=custom_palette)
plt.title('Distribution of Job Roles', fontsize=30)
plt.xlabel('Count')
plt.ylabel('Job Role')

plt.tight_layout()
plt.show()
```

###### 12.No.of Companies Worked

```python
sns.histplot(data=df,x='NumCompaniesWorked',bins=10,kde=True)
plt.title('Number Of Companies Worked')
plt.show()
```

###### 13.Distance from home

```python
plt.title('Distance From Home')
sns.histplot(data=df, x='DistanceFromHome', bins=10, kde=True)
plt.show()
```

###### 14.Overtime

```python
Overtime_Votes = df['OverTime'].value_counts()

custom_colors = ['#337357', '#FFD23F']

plt.figure(figsize=(8, 8))
plt.pie(Overtime_Votes, labels=Overtime_Votes.index, autopct=lambda p: f'{p:.1f}% ({int(p)})', colors=custom_colors)
plt.title('Count of Votes for Overtime', fontsize=30)
plt.axis('equal')
plt.show()
```
